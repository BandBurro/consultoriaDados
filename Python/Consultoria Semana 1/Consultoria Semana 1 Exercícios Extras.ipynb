# [1028. Figurinhas](https://judge.beecrowd.com/pt/problems/view/1028)

Entrada

A primeira linha da entrada contém um único inteiro N (1 ≤ N ≤ 3000), indicando o número de casos de teste. Cada caso de teste contém 2 inteiros F1 (1 ≤ F1 ≤ 1000) e F2 (1 ≤ F2 ≤ 1000) indicando, respectivamente, a quantidade de figurinhas que Ricardo e Vicente têm para trocar.

Saída

Para cada caso de teste de entrada haverá um valor na saída, representando o tamanho máximo da pilha de figurinhas que poderia ser trocada entre dois jogadores.

```python
import math
testes = int(input('numero de casos de teste '))
contador = 0
if testes >= 1 and testes <= 3000:
  while contador <= (testes-1):
    contador += 1
    fig1 = int(input('numero de figurinhas do primeiro '))
    if fig1 >= 1 and fig1 <= 1000:
      print(f'o primeiro jogador tem {fig1} figurinhas')
    fig2 = int(input('numero de figurinhas do segundo '))
    if fig2 >= 1 and fig2 <= 1000:
      print(f'o segundo jogador tem {fig2} figurinhas')
print('o tamanho máximo da pilha de figurinhas que poderia ser trocada entre dois jogadores é de',math.gcd(fig1, fig2))
```

# [1029. Fibonacci, Quantas Chamadas?](https://judge.beecrowd.com/pt/problems/view/1029)

Entrada
A primeira linha da entrada contém um único inteiro N, indicando o número de casos de teste. Cada caso de teste contém um inteiro X (1 ≤ X ≤ 39) .

Saída
Para cada caso de teste de entrada deverá ser apresentada uma linha de saída, no seguinte formato: fib(n) = num_calls calls = result, aonde num_calls é o número de chamadas recursivas, tendo sempre um espaço antes e depois do sinal de igualdade, conforme o exemplo abaixo.

```python
def fib(x, contagem):
    if x == 0:
        contagem['chamadas'] += 1
        return 0
    elif x == 1:
        contagem['chamadas'] += 1
        return 1
    else:
        contagem['chamadas'] += 1
        return fib(x-1, contagem) + fib(x-2, contagem)

N = int(input())
for _ in range(N):
    X = int(input())
    contagem = {'chamadas': 0}
    result = fib(X, contagem)
    print(f"fib({X}) = {contagem['chamadas'] - 1} cachamadas = {result}")
```

# [1035. Teste de Seleção 1](https://judge.beecrowd.com/pt/problems/view/1035)

Leia 4 valores inteiros A, B, C e D. A seguir, se B for maior do que C e se D for maior do que A, e a soma de C com D for maior que a soma de A e B e se C e D, ambos, forem positivos e se a variável A for par escrever a mensagem "Valores aceitos", senão escrever "Valores nao aceitos".

Entrada
Quatro números inteiros A, B, C e D.

Saída
Mostre a respectiva mensagem após a validação dos valores.

```python
lista = list()
contador = 0
while contador <= 3:
  contador += 1
  lista.append(int(input('digite um valor inteiro qualquer: ')))
vA, vB, vC, vD = lista
if vB > vC and vD > vA:
  if (vC + vD) > (vA+ vB):
    if vC and vD >= 0:
      if vA % 2 == 0:
        print('Valores aceitos')
      else:
        print('Valores não aceitos')
```
