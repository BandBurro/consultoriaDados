# Desafio 1 – Soma de Números Pares

- Exemplo de entrada: [1, 2, 3, 4, 5, 6]
- Saída esperada: 12

```python
lista = list()
while True:
  userinput = int(input('digite um valor inteiro qualquer e digite -1 para sair '))
  if userinput == -1:
    print(sum(lista))
    break
  else:
    if userinput % 2 == 0:
      lista.append(userinput)
```

# Desafio 2 – Filtro de Palavras

Objetivo: Dada uma lista de palavras, imprima apenas as que começam com a letra "a" ou "A".

- Exemplo de entrada: ["abacaxi", "Banana", "amora", "caju"]

- Saída esperada: abacaxi, amora

Note: não conhecia a função .startswith


```python
lista_polida = list()
while True:
  userinput = input('digite uma palavra qualquer ou sair para quebrar o loop ')
  if userinput.startswith(('a', 'A')):
    lista_polida.append(userinput)
  elif userinput == 'sair':
    break
print(lista_polida)
```

```python
lista = ["abacaxi", "Banana", "amora", "caju"]
for palavra in lista:
  if palavra.startswith(('a', 'A')):
    print(palavra)
```

 # Desafio 3 – Maior e Menor Número

 Objetivo: Peça ao usuário para digitar 5 números e diga qual é o maior e o menor valor.

- Entrada: 5 números via input
- Saída: O maior número é X, o menor é Y


```python
lista = list()
contador = 0
while contador <= 4:
  userinput = int(input('digite um dos cinco valores: '))
  lista.append(userinput)
  contador += 1
  print('Seu menor numero é:', min(lista), 'e seu maior numero é:', max(lista))
```

# Desafio 4 – Lista Invertida

Objetivo: Crie uma lista com 4 itens inseridos pelo usuário e imprima a lista na ordem inversa.

Exemplo:
- Entrada: "café", "leite", "água", "chá"
- Saída: ['chá', 'água', 'leite', 'café']

```python
lista = list()
contador = 0
while contador <= 3:
  contador += 1
  userinput = input('digite qualquer coisa: ')
  lista.append(userinput)
lista.reverse()
print(lista)
```

# Desafio 5 – Mini Calculadora

Objetivo: Peça dois números ao usuário e um operador (+, -, *, /), e execute a operação correspondente.
- Entrada:
- Número 1: 10
- Número 2: 5
- Operador: *
- Saída: Resultado: 50


```python
operação = input('escolha uma das operações: +, -, *, / ')
valor1 = float(input('digite o primeiro valor: '))
valor2 = float(input('digite o segundo valor: '))
if operação == '+':
 print(valor1 + valor2)
elif operação == '-':
  print(valor1 - valor2)
elif operação == '*':
  print(valor1 * valor2)
elif operação == '/':
  print(valor1 / valor2)
else:
  print('Operação inválida')
```
