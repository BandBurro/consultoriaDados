```python
def positivo_nulo_negativo(numero):
  if numero > 0:
    print("Valor Positivo")
  elif numero == 0:
    print("Valor nulo")
  else:
    print("Valor negativo")

# main
numero_digitado = float(input("Digite um número: "))
positivo_nulo_negativo(numero_digitado)
```

```python
def valor_absoluto(numero):
  if numero < 0:
    return numero * -1
  else:
    return numero

# main
numero_digitado = float(input("Digite um número: "))
valor_modulo = valor_absoluto(numero_digitado)
print(f"O valor absoluto de {numero_digitado} é {valor_modulo}")
```

```python
def adicao(a, b):
  resultado = a + b
  print(f"O resultado da adição é: {resultado}")

def subtracao(a, b):
  resultado = a - b
  print(f"O resultado da subtração é: {resultado}")

def multiplicacao(a, b):
  resultado = a * b
  print(f"O resultado da multiplicação é: {resultado}")

def divisao(a, b):
  if b != 0:
    resultado = a / b
    print(f"O resultado da divisão é: {resultado}")
  else:
    print("Erro: Divisão por zero!")

def main():
  operacao = input("Digite a operação desejada (+, -, x, /): ")
  valor1 = float(input("Digite o primeiro valor: "))
  valor2 = float(input("Digite o segundo valor: "))

  if operacao == '+':
    adicao(valor1, valor2)
  elif operacao == '-':
    subtracao(valor1, valor2)
  elif operacao == 'x':
    multiplicacao(valor1, valor2)
  elif operacao == '/':
    divisao(valor1, valor2)
  else:
    print("Operação inválida.")

main()
```

```python
def fatorial(n):
  if n < 0:
    return
  elif n == 0:
    return 1
  else:
    resultado = 1
    for i in range(1, n + 1):
      resultado *= i
    return resultado

# main
numero_lido = int(input("Digite um número inteiro para calcular o fatorial: "))
resultado_fatorial = fatorial(numero_lido)
print(f"O fatorial de {numero_lido} é {resultado_fatorial}")
```
