# Exercicio 1

Problema:
- Elabore o programa que leia um número qualquer e verifique se ele é
positivo, nulo ou negativo.

- Analise o problema e verifique quais são os dados que o usuário precisa
fornecer (digitar) como entrada.

```python
numero = float(input('digite um numero: '))
if numero > 0:
  print(f'{numero} é positivo ({numero*2})')
elif numero < 0:
  print(f'{numero} é negativo ({numero*3})')
else:
  print(f'{numero} é nulo')
```

# Exercicio 2

Problema:
- Construa o programa que leia dois valores quaisquer e mostre o maior
deles ou mostre a mensagem “Os valores são iguais.”

- Analise o problema e verifique quais são os dados que o usuário precisa
fornecer (digitar) como entrada.

```python
valor1 = float(input('digite o primeiro valor: '))
valor2 = float(input('digite o segundo valor: '))

if valor1 > valor2:
  print(f'o primeiro valor ({valor1}) é maior que o segundo ({valor2})')
elif valor2 > valor1:
  print(f'o segundo valor ({valor2}) é maior que o primeiro ({valor1})')
else:
  print(f'os valores {valor2} e {valor1} são iguais')
```

Alterações:

- Se eles forem diferentes, mostre os valores digitados na ordem decrescente
- Se eles forem iguais, mostre a mensagem e o valor digitado.

```python
valor1 = float(input('digite o primeiro valor: '))
valor2 = float(input('digite o segundo valor: '))

if valor1 > valor2:
  print(f'o primeiro valor ({valor1}) é maior que o segundo ({valor2})')
  print(f'{valor1}, {valor2}')
elif valor2 > valor1:
  print(f'o segundo valor ({valor2}) é maior que o primeiro ({valor1})')
  print(f'{valor2}, {valor1}')
else:
  print(f'os valores {valor2} e {valor1} são iguais')
```
