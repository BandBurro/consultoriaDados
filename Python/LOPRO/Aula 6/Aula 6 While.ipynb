Programa que leia varios numeros digitados pelo usuario e uso o valor -1 como condição (flag) de saida da estrutura de repetição.

Na tela de saida, mostre a quantidade de numeros digitados.

```python
ct = 0
print('digite -1 para sair da repetição')
while True:
  numero = int(input('Digite um numero'))
  if numero == -1:
    break
  ct = ct + 1
print('quantidade de numeros digitados:', ct)
```

Desenvolva o programa que calcule a media aritmetica de uma turma, onde cada launo realizou uma avaliação. Nao sabemos a quantidade de alunos, por isso, usaremos o valor "-1" como condição (flag) de saida

Na tela de saida, mostre a media aritimetica da turma e a quantidade de alunos na turma.

Analise o problema e verifique quais sao os dados que o usuario precisa fornecer como entrada.

```python
ct = 0
somador = 0
print('digite -1 pra sair')
while True:
  numero = float(input('digite um numero'))
  if numero == -1:
    break
  somador = somador + numero
  ct = ct + 1
print('numero de alunos é:', ct)
print('media é:', somador/ct)
```

para fazer um contador em loop é imprescindivel ter a estrutura x = x + y em variaveis

Alterações:
1. Mostre tambem a soma das notas
2. Mostre a media da turma com duas casas decimais
3. No inicio do programa leia o nome da disciplina
4. E no final mostre tambem a disciplina de cada nota

```python
ct = 0
somador = 0
print('digite -1 pra sair: ')
while True:
  numero = float(input('digite um numero: '))
  if numero == -1:
    break
  disciplina = input('digite a disciplina: ')
  somador = somador + numero
  ct = ct + 1
print('numero de alunos é: ', ct)
print(f'media é:{somador/ct: .2f}')
print('soma das notas é: ', somador)
print('disciplina: ', disciplina)
```

- Construa o programa que calcule a média aritmética dos números pares.
O usuário fornecerá os valores de entrada que pode ser um número qualquer par ou impar. A condição de saída será o número 0 (zero).
Analise o problema e verifique quais são os dados que o usuário precisa
fornecer (digitar) como entrada.
Passos para resolução do problema (algoritmo):

Dicas: 1. media soma / ct
2. 0 operador modulo (%) pega o resto da divisão de dois números.
Teste 1: Entrada: valor: 1, 2 e 0
Saida: Média 2
Teste 2: Entrada: valor: 1, 2, 3, 4 e 0
Saída: Média 3
Teste 3:
Entrada: valor: 1, 2, 2, 4 e 0
Saida: Media 2.66666666666666666665

```python
ct = 0
par = 0
print('digite um numero ')
while True:
  numero = int(input('digite um numero '))
  if numero == 0:
    break
  elif (numero%2) == 0:
    par = par + numero
    ct = ct + 1

print('media dos pares é: ', (par/ct))
```
