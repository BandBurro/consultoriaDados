```python
v_menor = 9999999999999
v_maior = -9999999999999
ct = 0
somador = 0
print('digite 0 para sair da repeitção')
while True:
  valor = float(input('digite seu valor: '))
  if valor == 0:
    break
  if valor < v_menor:
    v_menor = valor
  if  valor > v_maior:
    v_maior = valor
  somador += valor
  ct += 1
print(f'{valor}')
print(f'{v_maior}')
print(f'{v_menor}')
print(f'{somador}')
print(f'{somador/ct}')
```
