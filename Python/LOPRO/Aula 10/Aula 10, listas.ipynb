```python
# indice de listas
lista1 = [3, 'abacate', 9.7, [5, 6, 3], "Python", (3, 'j')]
print(lista1[5])
print(lista1[3][1]) # leu o valor 3 da lista1, ou seja, a lista principal, e o valor 1 desta mesma lista interna, lembrando que a contagem começa do 0

# criação de listas
lista_um = [ ] # lista sem nada
lista_um = list()
```

```python
# CRUD de uma lista utiliza os comandos create, read, update e delete
# create
lista = [1,2,3,4,5,6,7,8,9]
lista.append(2) #valor aleatorio, esta função adiciona um elemento, qualquer que seja, int, float ou string, no FINAL da lista
print(lista)

lista1 = [1,2,3,4,5,6,7,8]
lista1.insert(0,3) # (i,x)
lista1.insert(0,7) # (i,x)
# Insere o item ‘x’ na posição ou índice ‘i’ e desloca os valores para DIREITA.
# O primeiro argumento é o índice do elemento (i) antes do qual será feita a inserção.
print(lista1) # empurrou tudo para direita a partir do i (1), e adicionou o valor explicitado (3)
```

```python
# read, ou consulta
lista = [1,2,3,7,8,11,54,123]
print(lista, 'lista') # Consulta e mostra todos os valores armazenados na lista na horizontal.

for i in lista: # Consulta e mostra todos os valores armazenados na lista na horizontal,
  print(lista)

for i in lista: # Consulta e mostra todos os valores armazenados na lista na vertical,
  print(i)

boolean = 5 in lista # O in verifica se o valor ’x’ pertence à lista. Retorna True ou False
print(boolean)
if 5 in lista:
  print('ta la')
else:
  print('ta nao')

test = lista.index(3, 0, 4) # acha a posição (indice) do valor digitado, ele deve existir na lista
print(test)
# . Retorna o índice do primeiro item cujo valor é igual a ‘x’.
# . Os argumentos opcionais start (início) e end (fim) são interpretados como na slice notation e são usados para limitar a pesquisa a uma subsequência específica da lista.
# . O índice retornado é calculado em relação ao início da sequência completa em vez do argumento inicial.
```

```python
# update
lista1 = [3, 'abacate', 9.7, [5, 6, 3], "Python", (3, 'j')]
lista1 [3] = 'morango' # O valor da posição ‘i’ (3) será substituído pelo novo_valor (morango)
print(lista1)

# A tentativa de acesso a um índice inexistente resulta em erro. e.g lista1 [7]
```

```python
# delete
lista1 = [3, 'abacate', 9.7, [5, 6, 3], "Python", (3, 'j')]
lista1.remove('abacate') # Remove o primeiro item encontrado na lista com valor = ‘x’ EXATO, não seu indice, e desloca os valores para esquerda.
print(lista1) # Removeu o primeiro 3
```

```python
# ordenar
lista = [30, 20, 70, 40, 50]
# Ordena os elementos da lista in-place, ou seja, modifica a lista original no mesmo lugar na memória, ou seja, altera a lista por definitivo
lista.sort(reverse=False) #
print('lista.sort(reverse=False)', lista)
lista.sort(reverse=True)
print('lista.sort(reverse=True)', lista)
lista.sort()
print('lista.sort()', lista)
print()
print()
print()
# sorted
# Retorna os elementos da lista em uma ordem crescente e não altera a lista original
lista1 = [3, 4, 1, 5, 2]
nova_lista = sorted(lista1)
print(nova_lista)
print(lista1)
print()
print()
print()
# reverse
# Inverta os elementos da lista, não coloca os dados em ordem decrescente. Altera a lista original e retorna None.
lista2 = [1, 2, 7, 3, 4, 5]
lista2.reverse()
print(lista2)
print()
print()
print()
# Não altera a lista original, retorna o resultado na nova lista
lista3 = [1, 2, 3, 4, 5]
nova_lista2 = list(reversed(lista3))
print(nova_lista2)
```

```python
# numerico
# max(lista) Retorna o maior elemento da lista
lista = [1, 2, 3, 4, 5]
m = max(lista)
print(m)
# min(lista) Retorna o menor elemento da lista
mn = min(lista)
print(mn)
# sum(lista) Retorna a soma dos elementos da lista
soma = sum(lista)
print(soma)
# len(lista) Retorna a quantidade de elementos da lista
qtd = len(lista)
print(qtd)
# lista.count(x) Retorna o número de vezes que o item ‘x’ aparece na lista
lista2 = [1, 2, 2, 2, 3, 4, 5]
print(lista2.count(2)) # o 2 aparece 3 vezes
print(lista2.count(6)) # o 6 aparece 0 vezes
```

| Função  | Descrição                                       | Exemplo                                                       |
|---------|-------------------------------------------------|----------------------------------------------------------------|
| `len`   | Retorna o tamanho da lista.                     | `li = [1, 2, 3, 4]` → `len(li)` resulta em `4`                |
| `count` | Retorna quantas vezes x aparece na lista.       | `li = [1, 2, 3, 3, 4]` → `li.count(3)` resulta em `2`         |
| `min`   | Retorna o menor valor da lista.                 | `li = [10, 40, 30, 20]` → `min(li)` resulta em `10`           |
| `max`   | Retorna o maior valor da lista.                 | `li = [10, 40, 30, 20]` → `max(li)` resulta em `40`           |
| `sum`   | Retorna a soma dos elementos da lista.          | `li = [10, 20, 30]` → `sum(li)` resulta em `60`               |
| `append`| Adiciona um novo valor no final da lista.       | `li = [1, 2, 3]` → `li.append(100)` → `li = [1, 2, 3, 100]`    |
| `in`    | Verifica se um valor pertence à lista.          | `li = [1, 2, 3, 4]` → `3 in li` resulta em `True`             |
| `sort()`| Ordena a lista em ordem crescente.              | `li = [3, 5, 2, 4, 1, 0]` → `li.sort()` → `li = [0, 1, 2, 3, 4, 5]` |
